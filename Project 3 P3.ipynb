```python
basics = pd.read_csv('Data/title_basics.csv.gz')
basics
```

```python
ratings = pd.read_csv('Data/title_ratings.csv.gz')
ratings
```

```python
basics['genres_split'] = basics['genres'].str.split(',')
basics
```

```python
unique_genres = basics['genres_split'].explode().unique
unique_genres
```

```python
exploded_genres = basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
title_genres = exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
genre_id_map = dict(zip(unique_genres,range(len(unique_genres))))
genre_id_map
```

```python
## make new integer genre_id and drop string genres
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns='genres_split')
```

```python
genre_lookup = pd.DataFrame({
'Genre_Name': genre_id_map.keys(),
'Genre_ID': genre_id_map.values()})

genre_lookup.head()
```

```python
basics.drop(columns=['originalTitle', 'isAdult','titleType','genres','genres_split'],inplace=True)
basics.head()
```

```python
basics.dtypes
```

```python
#Basics schema creation and table creation.

key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}

# Save to sql with dtype and index=False
basics.to_sql('title_basics',engine,dtype=df_schema,if_exists='replace',index=False)

engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

```python
ratings.dtypes
```

```python
#Ratings schema creation and table creation.

key_len = basics['tconst'].fillna('').map(len).max()
title_len = basics['primaryTitle'].fillna('').map(len).max()

df_schema = {
    "tconst": String(key_len+1), 
    'averageRating':Float(),
    'numVotes':Integer()}

ratings.to_sql('title_ratings',engine,dtype=df_schema,if_exists='replace',index=False)

engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')
```
